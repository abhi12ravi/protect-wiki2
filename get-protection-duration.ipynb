# Define helper functions

```python
#Get current protection status of article 

def get_protection_status(title):
    
    import requests
    S = requests.Session()

    URL = "https://en.wikipedia.org/w/api.php"

    PARAMS = {
        "action": "query",
        "format": "json",
        "titles": title,
    #     "list": "protectedtitles",
        "prop": "info",
        "inprop": "protection"
    }

    R = S.get(url=URL, params=PARAMS)
    DATA = R.json()

    PAGES = DATA["query"]["pages"]
    
    for k, v in PAGES.items():
        if v['protection'] != []:
            protection_status = v["protection"][0]['level'] 
            protection_expiry = v["protection"][0]['expiry']
            print('Protection status = ',protection_status)
            print('Protection duration =', protection_expiry)
            return protection_status
```

```python
title = 'Best Korea',
protection_status = get_protection_status(title)
print("Title: ", title, "Protection Status:", protection_status) 
```
